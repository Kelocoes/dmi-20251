```python
import pandas as pd

df = pd.read_csv('chart-data.csv')
df.shape
```

```python
df.head()
```

```python


# Convert 'Date' and 'Time' columns to datetime
df['Time'] = df['Time'].str.replace('a. m.', 'AM').str.replace('p. m.', 'PM')
df['DateTime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], format='%d/%m/%Y %I:%M:%S %p')

# Sort the DataFrame by the 'DateTime' column
df = df.sort_values('DateTime')

# Calculate the time difference in seconds
df['TimeDiff'] = df['DateTime'].diff().dt.total_seconds()

# Save the differences to a txt file
with open('diffs.txt', 'w') as f:
    for i in range(1, len(df)):
        start_time = df['DateTime'].iloc[i-1]
        end_time = df['DateTime'].iloc[i]
        time_diff = df['TimeDiff'].iloc[i]
        # if (time_diff >  10):
        f.write(f'Difference between {start_time} and {end_time}: {time_diff} seconds\n')
```
